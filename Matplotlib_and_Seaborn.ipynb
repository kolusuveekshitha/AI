```python
sns.scatterplot(x=x,y=y)
```

## Pie chart

```python
data=[2,6,8,10]
plt.pie(data)
plt.show()
```

```python
plt.pie(data,[0,0,.09,0],labels=['t1','t2','t3','t4'],autopct="%1.1f%%",shadow=True,colors=['Pink','Blue','Violet','orange'])
plt.title('Pie Diagram')
plt.show()
```
