## Merge

```python
dd=pd.DataFrame({'Name':['Veekshitha','ishu','ananda','kavya'],'English':[92,93,94,94]})
dd
```

```python
dm=pd.DataFrame({'Name':['Veekshitha','ishu','ananda','kavya'],'Maths':[96,91,95,91]})
dm
```

```python
m1=pd.merge(dd,dm,on='Name')
m1
```

```python
dh=pd.DataFrame({'Name':['Veekshitha','ishu','ananda','kavya'],'history':[98,99,90,94]})
dh
```

```python
m2=pd.merge(m1,dh)
m2
```
