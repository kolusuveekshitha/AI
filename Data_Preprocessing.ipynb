## Outlier Treatment - Replacing outliers with median

```python
# Upper_limit = q3+1.5*IQR

# Lower_limit = q1-1.5*IQR
```

```python
df.describe()
```

```python
q1=df.bmi.quantile(0.25)
q1
```

```python
q3=df.bmi.quantile(0.75)
q3
```

```python
IQR=q3-q1
IQR
```

```python
Upper_limit = q3+1.5*IQR
Upper_limit
```

```python
Lower_limit = q1-1.5*IQR
Lower_limit
```

```python
df.bmi.median()
```

```python
df['bmi']=np.where(df['bmi']>Upper_limit,30,df['bmi'])
```

```python
sns.boxplot(df.bmi)
```
