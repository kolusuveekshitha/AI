## Multi Linear Regression

```python
df=pd.read_csv('/content/50_Startups.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.STATE.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.STATE=le.fit_transform(df.STATE)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.corr()
```

```python
import seaborn as sns
sns.heatmap(df.corr(),annot=True)
```

```python
df.corr().PROFIT.sort_values(ascending=False)
```

```python
# x and y split

x=df.drop(columns=['PROFIT'],axis=1)
x.head()
```

```python
y=df.PROFIT
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_scaled=pd.DataFrame(sc.fit_transform(x),columns=x.columns)
x_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.3,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
y_test
```

```python
salary=pd.DataFrame({'Actual_profit':y_test,'Predicted_profit':y_pred})
salary
```

```python
df.head()
```

```python
# testing with random value
lr.predict([[165349.20,101145.10,383199.62,0]])/100000
```

## Evaluation Metrics for Regression

```python
from sklearn import metrics
```

```python
# R-square
# on test data
metrics.r2_score(y_test,y_pred)
```

```python
y_pred1=lr.predict(x_train)
y_pred1
```

```python
# on training data

metrics.r2_score(y_train,y_pred1)
```

```python
# MSE-Mean Square Error

metrics.mean_squared_error(y_test,y_pred)
```

```python
# RMSE - Root Mean Square Error
import numpy as np
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
